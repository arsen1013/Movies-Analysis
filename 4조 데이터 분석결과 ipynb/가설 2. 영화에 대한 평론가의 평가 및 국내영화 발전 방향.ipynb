수집한 내용 전처리

```python
import re
def preprocess(text):
  text = re.sub('[-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·]', '', text)
  text = re.sub('[a-zA-z]','',text)
  return(text)
```

```python
clean1=preprocess(str(frame1))
clean2=preprocess(str(frame2))
clean3=preprocess(str(frame3))
clean4=preprocess(str(frame4))
clean5=preprocess(str(frame5))
```

워드클라우드 생성

```python
from konlpy.tag import Okt

okt=Okt()
okt1 = okt.nouns(clean1)
okt2 = okt.nouns(clean2)
okt3 = okt.nouns(clean3)
okt4 = okt.nouns(clean4)
okt5 = okt.nouns(clean5)
```

```python
from wordcloud import WordCloud
from collections import Counter
```

```python
count1 = Counter(okt1)
data1 = count1.most_common(300)

count2 = Counter(okt2)
data2 = count2.most_common(300)

count3 = Counter(okt3)
data3 = count3.most_common(300)

count4 = Counter(okt4)
data4 = count4.most_common(300)

count5 = Counter(okt5)
data5 = count5.most_common(300)
```

```python
#불용어 생성

stopwords_kr = ['하지만', '그리고', '그런데', '저는','제가',  '있다', 
                '그럼', '이런', '저런', '합니다', '있습니다',
                '많은', '많이', '정말', '너무'] 
```

```python
def displayWordCloud(data = None, backgroundcolor = 'white', width=800, height=600 ):
    wordcloud = WordCloud(
                        font_path = '/Library/Fonts/NanumBarunGothic.ttf', 
                        stopwords = stopwords_kr,
                        background_color = backgroundcolor, 
                         width = width, height = height).generate(data)
    plt.figure(figsize = (15 , 10))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show() 
```

```python
result1 = {}
for a,b in data1:
  result1[a]=b
```

```python
displayWordCloud(' '.join(result1))
```
